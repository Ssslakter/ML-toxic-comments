```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/train.csv")
df_test = pd.read_csv("../data/test.csv")
```

```python
df['comment_text'].apply(lambda x: len(x.split(' '))
                         ).sort_values(ascending=False).head(30)
```

```python
plt.figure(figsize=(15, 5))
plt.hist(df['comment_text'].apply(lambda x: len(x.split(' '))), bins=100)
```

```python
toxic = df[df['toxic'] == 1]
not_toxic = df[df['toxic'] == 0]
fig, ax = plt.subplots(2, 1)
ax[0].hist(toxic['comment_text'].apply(lambda x: len(x.split(' '))), bins=1000)
ax[1].hist(not_toxic['comment_text'].apply(
    lambda x: len(x.split(' '))), bins=1000)
```
